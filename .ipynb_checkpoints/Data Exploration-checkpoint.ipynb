```python
def clean_info(df):
    pd.set_option('display.max_colwidth', -1) 
    df['Year'] = df['Info'].str[8:12]
    df = df[df['Year'].str.contains("19")]
    df['Year'] = df['Year'].astype(int)
    df = df[(df['Year'] >= 1926) & (df['Year'] < 1960)]
    
    df['Artist Info'] = df['Info'].str[13:]
    df['MP3 Avaliable'] = 'Online'
    # Find multi-links:
    df['comma_loc'] = df['Download_link'].str.find(",")
    df['Link'] = np.where(df['comma_loc'] == -1, df['Download_link'], df['Download_link'].str[:47])
    df['Artist'] = df['Artist'].str.strip()
    
    return df[['Artist', 'Title', 'Year', 'MP3 Avaliable', 'Link']]
    
```

```python
##Load & Clean Download CSV

csv_file = 'music_downloads/DownloadA.csv'

def clean_download_list(csv_file):
    
    #Load the file
    df_download = pd.read_csv(csv_file)

    #Create unique filenames for mp3 (ID & Style)
    df_download['Style-Katie'] = df_download['Style-Katie'].replace("Lindy Hop", "Lindy")
    df_download['filename'] = (df_download['ID'].astype(str)
                                            .str.zfill(4)
                                            + '-' +
                                            df_download['Style-Katie'])
    #Drop duplicate filepaths (keep last)
    df_download.drop_duplicates(subset='Link', keep='last', inplace=True)
    
    return df_download

```

```python
import urllib2



def download_mp3(df, start, stop):
    
    if stop-start > 50:
        print "batch is too large"
    else:
        for mp3, name in zip(df['Link'][start:stop], df['filename'][start:stop]):
            try:
                mp3file = urllib2.urlopen(mp3)
                with open('music_downloads/{}.mp3'.format(name),'wb') as output:
                    output.write(mp3file.read())
                #Pause between 3rd of a second to 3 and a half seconds
                pause.milliseconds(random.randint(300, 3500)) 
            except:
                print "{} is not valid :  {}".format(mp3, name) 
                with open('error.txt', 'a') as f:
                    f.write('{},{},{}\n'.format(name[:4],mp3,name))
```

```python
start = 71
stop = 83

download_mp3(df_download, start, stop)
```
